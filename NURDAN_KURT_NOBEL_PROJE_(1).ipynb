```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df["birth_date"]=pd.to_datetime(df["birth_date"])
df["year"]=pd.to_datetime(df["year"])
df["age"]=df["year"].dt.year-df["birth_date"].dt.year
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x="year", y="age", hue="sex", style="category")
plt.title("Age Distribution of Nobel Prize Winners by Year and Sex")
plt.xlabel("Year")
plt.ylabel("Age")
plt.legend(title="Sex")
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(True)
plt.show()
```

6. ANALYSIS

```python
df["year"] = pd.to_datetime(df["year"], format="%Y")
filtered_years=df[(df["year"].dt.year >= 1938) & (df["year"].dt.year <= 1945)]
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=filtered_years, x="category", palette="viridis")
plt.title("Distribution of Nobel Prize Winners Between 1938 and 1945")
plt.xlabel("Category")
plt.ylabel("Count of Winners")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12,6))
sns.countplot(data=filtered_years, x="organization_country", palette="magma", order=filtered_years["organization_country"].value_counts().index[:10])
plt.title("Distribution of Nobel Prize Winners Between 1938 and 1945")
plt.xlabel("Country")
plt.ylabel("Count of Winners")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
df["year"] = pd.to_datetime(df["year"], format="%Y")
selected_categories = ["Chemistry", "Physics", "Literature", "Medicine", "Peace"]
filtered_data=df[(df["year"].dt.year >= 2000) & (df["category"].isin(selected_categories))]
filtered_data=filtered_data.dropna(subset=["birth_date"])
```

```python
filtered_data["birth_date"]=pd.to_datetime(filtered_data["birth_date"], errors='coerce')
filtered_data["age_at_award"]=filtered_data["year"].dt.year-filtered_data["birth_date"].dt.year
```

```python

colors=sns.color_palette("Set2")
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=filtered_data, x="category", y="age_at_award", palette="Set2")
plt.title("Age Distribution of Nobel Prize Winners After 2000 by Category")
plt.xlabel("Category")
plt.ylabel("Age at Award")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
